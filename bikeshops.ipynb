```python
# Charger le fichier pour avoir un aperçu des premières lignes
df_bikeshops = pd.read_excel("C:/Users/Jalal/Desktop/Projet_MrSelami/data/bikeshops.xlsx")
df_bikeshops.head()
```

```python
# Afficher les premières lignes pour vérifier l'importation
print(df_bikeshops.head())

# Statistiques descriptives pour une colonne numérique pertinente
# Remplacer 'nom_colonne_numerique' par le nom de la colonne numérique pour laquelle vous voulez des statistiques
print(df_bikeshops['bikeshop.id'].describe())

# Calcul de l'écart type pour la même colonne
print(df_bikeshops['bikeshop.id'].std())

# Compter le nombre d'occurrences pour une colonne catégorielle
# Remplacer 'nom_colonne_categorique' par le nom de la colonne catégorielle
print(df_bikeshops['bikeshop.city'].value_counts())

# Prix (ou autre mesure) moyen par catégorie
# Remplacer 'nom_colonne_categorique' et 'nom_colonne_numerique' par les noms réels
print(df_bikeshops.groupby('bikeshop.city')['bikeshop.id'].mean())
```
